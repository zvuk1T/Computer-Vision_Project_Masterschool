```python
import cv2
```

```python
img = cv2.imread('/Users/zeal.v/Desktop/Computer-Vision_Project_Masterschool/DATA/viber_image_2025-02-03_10-08-53-621.jpg')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
imageLine = img_rgb.copy()
pointA = (200, 80)
pointB = (450, 80)

cv2.line(imageLine, pointA, pointB, (0, 255, 255),thickness = 3, lineType = cv2.LINE_AA)
plt.imshow(imageLine)
```

```python
imgCircle = img_rgb.copy()
circle_center = (415, 220)
radius = 200
cv2.circle(imgCircle, circle_center, radius, (255, 0, 0),
           thickness = 3,
           lineType = cv2.LINE_AA)
plt.imshow(imgCircle)
```

```python
imagefilledCircle = img_rgb.copy()
circle_center = (410, 230)
radius = 170

cv2.circle(imagefilledCircle
           , circle_center
           , radius, (0, 255, 0)
           , thickness = -1
           , lineType = cv2.LINE_AA)
plt.imshow(imagefilledCircle)
```

```python
imageRectangle = img_rgb.copy()
start_point = (285, 75)
end_point = (540, 400)

cv2.rectangle(imageRectangle,
              start_point,
              end_point,
              (0, 255, 0),  # Green color
              thickness=3,
              lineType=cv2.LINE_AA)

plt.imshow(imageRectangle)
```

```python
imageEllipse = img_rgb.copy()
center_coordinates = (410, 230)
axesLength = (200, 170)
angle = 90
startAngle = 0
endAngle = 360

# First ellipse
cv2.ellipse(imageEllipse, center_coordinates, axesLength,
            angle, startAngle, endAngle, (255, 255, 0),
            thickness=3, lineType=cv2.LINE_AA)

# Second ellipse rotated by 90 degrees
cv2.ellipse(imageEllipse, center_coordinates, axesLength,
            angle + 90, startAngle, endAngle, (0, 255, 255),
            thickness=3, lineType=cv2.LINE_AA)

plt.imshow(imageEllipse)
```

```python
imageText = img_rgb.copy()
text = 'Hello, Computer Vision!'
org = (50, 50)
cv2.putText(imageText, text, org, cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 0, 0), 2, cv2.LINE_AA)
plt.imshow(imageText)
```
